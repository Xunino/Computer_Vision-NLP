```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
# Sentences
sentences = [
             "Tôi yêu Việt Nam",
             "Hôm nay trời tuyệt đẹp",
             "Tôi là người giỏi nhất",
             "Tôi luôn làm những điều tuyệt vời"
             ]
```

```
# Căt từ chuẩn bị dữ liệu
# Giới hạn từ điển là 100 từ
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
```

```
# Xây dựng từ điển
tokenizer.fit_on_texts(sentences)
```

```
# Kiểm tra thư viện 
word_index = tokenizer.index_word
```

```
word_index
```

```
new_sentences = [
                 "I love HaNoi",
                 "Goodbye see you late"
]
```

```
new_sentences = tokenizer.texts_to_sequences(new_sentences)
```

```
# Kiểm tra câu trong từ điển ban đầu
new_sentences
```

```
# Thêm Padding - thêm một từ vào để tạo nên 1 các sequences có độ dài bằng nhau
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```
padding_sequences = pad_sequences(new_sentences)
```

```
print(padding_sequences)
```

```
# Chuyển câu thành ma trận thứ tự chữ trong từ điển
sentences = tokenizer.texts_to_sequences(sentences)
```
