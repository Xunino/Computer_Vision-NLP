```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
path = '/content/drive/My Drive/BT_ML/chess.png'
image = cv2.imread(path)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# print(image.shape[0]) 

# ????????
# print(len(image))
# print(len(image[0]))

plt.gray()
plt.imshow(gray)
```

```
def maxpooling(image, stride):
  width = image.shape[1]
  height = image.shape[0]

  # Lưu ảnh vào một ma trận rỗng hình vuông
  new_image = np.zeros((height // 2 + 1, width // 2 + 1))

  xRows = [1, 0, 1]
  yCols = [0, 1, 1]

  for i in range(0, height, stride):
    for j in range(0, width, stride):
      currentPixels = [image[i][j]]

      for k in range(3):
        neighbor_x = i + xRows[k]
        neighbor_y = j + yCols[k]

        if neighbor_x < height and neighbor_y < width:
          currentPixels.append(image[neighbor_x][neighbor_y])

        new_image[i//2][j//2] = max(currentPixels)
  return new_image
```

```
result = maxpooling(gray, 3)
plt.imshow(result)
```
